#### This python script is to combine the `track_name` and `track_artist` features from this dataset for better visualisation on our scatter plot, and then drop the old columns, and save the new dataset to the new `spot_songs_processed.csv`

```python
# Import needed libraries
import pandas as pd

# Replace 'path/to/spotify_songs.csv' with the actual path to your CSV file
file_path = 'spotify_songs.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first 5 rows of the DataFrame
df.head(5)
```

##### Before combining `track_name` with `track_artist`, lets firstly add "by " to the `track_artist` column

```python
df['track_artist'] = ' by ' + df['track_artist']
df.head(5)
```

##### Combine the `track_name` and `track_artist` columns

```python
df['track_name_and_artist'] = df['track_name'] + df['track_artist']
df.head(5)
```

#### Drop the old columns

```python
df = df.drop('track_artist', axis=1)
df = df.drop('track_name', axis=1)
df.head(5)
```

##### Move the new column to the front

```python
# Assuming 'track_name_and_artist' is the current last column
track_name_and_artist_column = df['track_name_and_artist']
df = df.drop('track_name_and_artist', axis=1)  # Remove the column from its current position
df.insert(1, 'track_name_and_artist', track_name_and_artist_column)  # Insert the column at the desired position
df.head(5)
```

```python
df = df.sample(frac=0.10, random_state=42)  # Set random_state for reproducibility
```

#### We also want to map the `key` values to their corresponding key represented by characters:

```python
def map_key_to_string(key_value):
    if key_value == -1:
        return "No key detected"
    elif key_value == 0:
        return "C"
    elif key_value == 1:
        return "C#/Db"
    elif key_value == 2:
        return "D"
    elif key_value == 3:
        return "D#/Eb"
    elif key_value == 4:
        return "E"
    elif key_value == 5:
        return "F"
    elif key_value == 6:
        return "F#/Gb"
    elif key_value == 7:
        return "G"
    elif key_value == 8:
        return "G#/Ab"
    elif key_value == 9:
        return "A"
    elif key_value == 10:
        return "A#/Bb"
    elif key_value == 11:
        return "B"
    else:
        return "Invalid key value"

df['key'] = df['key'].apply(lambda x: map_key_to_string(x))
df.head(5)
```

##### Save the new dataset to `spotify_songs_processed.csv`

```python
# Save the DataFrame to a new CSV file
df.to_csv('spotify_songs_processed.csv', index=False)
```
